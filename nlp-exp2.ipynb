## BOW using dictionary method .get()

```python
text_data=[]
for word in car_data2:
        for w in word:
            if re.findall("(\w*)([^0-9])",w):
                text_data.append(w)
text_data
```

```python
BOW={}
for word in car_data2:
    for w in word:
        if w in text_data:
            BOW[w]=BOW.get(w,0)+1

print(len(BOW))            
print(BOW)
```
